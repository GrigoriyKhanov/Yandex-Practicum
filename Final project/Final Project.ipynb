# Отток клиентов

### Описание проекта: телекоммуникации

Оператор связи «ТелеДом» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор.

Предоставлены  персональные данные о некоторых клиентах, информацию об их тарифах и услугах.
Необходимо обучить модель для прогноза оттока клиентов.

### Описание услуг:

Оператор предоставляет два основных типа услуг: 

    Стационарную телефонную связь. 
    Телефон можно подключить к нескольким линиям одновременно.

    Интернет.
    Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line — «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).

Также доступны такие услуги:

    Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);

    Выделенная линия технической поддержки (TechSupport);

    Облачное хранилище файлов для резервного копирования данных (OnlineBackup);

    Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).

Клиенты могут платить за услуги каждый месяц или заключить договор на 1–2 года. Возможно оплатить счёт разными способами, а также получить электронный чек.

```python
def auc_roc (features, target, model):
    model = model
    probabilities = model.predict_proba(features)
    probabilities_one = probabilities[:, 1]

    fpr, tpr, thresholds = roc_curve(target, probabilities_one) 
    auc_roc = roc_auc_score (target, probabilities_one)

    print("Значение AUC_ROC модели равно:", auc_roc)
    plt.figure(figsize=(10, 10))
    plt.plot(fpr, tpr, linestyle='-')
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-curve')
    plt.savefig('ROC-curve.png')
    plt.show()
```

# 6. Обучение моделей машинного обучения

Выполним обучение моделей для этого воспользуемся функцией pipeline.
С помощью ее выполним  кодирование данных и масштабирование данных  используя OneHotEncoder и StandardScaler.

```python
cat_columns = features_train.select_dtypes(include=['object', 'category']).columns.tolist()
cat_columns
```

```python
num_columns = features_train.select_dtypes(include=['int64',float]).columns.tolist()

num_columns
```

```python
scaler = StandardScaler( )

encoder = OneHotEncoder(drop="first")

num_pipeline = make_pipeline(scaler,)

cat_pipeline = make_pipeline(encoder)
                                      
column_transformer = make_column_transformer((num_pipeline, num_columns),
                                             (cat_pipeline, cat_columns),
                                              remainder='passthrough' 
                                             )                 
column_transformer
```

Для подбора параметров воспользуемся RandomizedSearchCV. В качестве первой модели примем модель случайного леса.

```python
RF_model = RandomForestClassifier(random_state=RANDOM_STATE)  

RF_parameters = {
   'randomforestclassifier__n_estimators': range(100, 1000, 50) ,
   'randomforestclassifier__max_depth': range(6, 12, 3),
   'randomforestclassifier__min_samples_split' : range(2, 20, 2) ,
   'randomforestclassifier__class_weight': ['balanced']
}
    
    
RF_pipeline = make_pipeline(column_transformer, RF_model) 
```

```python
random_search_RF = RandomizedSearchCV(RF_pipeline,
                                      param_distributions = RF_parameters ,
                                      n_iter= 200, 
                                      cv=3,
                                      n_jobs=-1, 
                                      scoring='roc_auc',
                                      random_state=RANDOM_STATE
                                     )
```

```python
%%time
random_search_RF.fit(features_train, target_train)
```

```python
random_search_RF.best_params_
```

```python
random_search_RF.best_score_
```

В качестве второй  модели рассмотри  модель семейства Boosting LightGBM LGBMClassifier

```python
LGBM_model = LGBMClassifier(random_state=RANDOM_STATE)
LGBM_parameters = {
   'lgbmclassifier__n_estimators': range(100, 1000, 50),
   'lgbmclassifier__max_depth': range(6, 12, 3),
   'lgbmclassifier__min_child_samples': range(20, 100, 10),
   'lgbmclassifier__num_leaves': range(20, 100, 10),
   'lgbmclassifier__reg_lambda': [0.2, 0.4, 0.6, 0.8, 1],
   'lgbmclassifier__class_weight': ['balanced']
   
}
   
    
LGBM_pipeline = make_pipeline(column_transformer, LGBM_model ) 
```

```python
random_search_LGBM = RandomizedSearchCV(LGBM_pipeline,
                                      param_distributions = LGBM_parameters ,
                                      n_iter= 200, 
                                      cv=3,
                                      n_jobs=-1, 
                                      scoring='roc_auc',
                                      random_state=RANDOM_STATE
                                     )
```

```python
%%time
random_search_LGBM.fit(features_train, target_train)
```

```python
random_search_LGBM.best_params_
```
